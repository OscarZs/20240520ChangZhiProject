```python
#读取原始od数据
import pandas as pd
import numpy as np
od = pd.read_excel('../data/OD_Demand_of_Changzhi_City _2024_A.xlsx',sheet_name='Sheet1')

#od.head()

# 计算小区出行量

tmp = od.drop('小区名称', axis=1)
tmp = tmp.sum()
tmp = tmp.reset_index()
tmp.columns = ['小区编号', '出行量']
tmp.to_csv('小区出行量.csv', index=False)
```

```python
# 读取od 数据
import pandas as pd

od_by_area = pd.read_csv('OD_By_Area.csv')

od_by_area.head()
```

```python
# 计算小区总出行量

# 小区名称对应了编号，用字典保存
area_index = {
    '城北科技园':1,
    '城北生态居住区':2,
    '长兴路居住区':3,
    '城北科技园服务区':4,
    '太行西街现代居住区':5,
    '高等教育园区':6,
    '五一路现代居住区':8,
    '西八一广场商务区 ':9,
    '东八一广场商务区':10,
    '城东现在居住区':11,
    '火车站物流商贸区':12,
    '火车站配套居住区':13,
    '解放西路居住区':14,
    '城南教育科研区':15,
    '解放东路居住区':16,
    '城南生态居住区':17,
    '生态区居住区':18,
    '产业配套居住区':19,
    '城区机械工业园区':20,
    '新区起步区':21,
    '老顶山开发区':22,
    '湖东北部新区':23,
    '湖西新区':24,
    '马厂':25,
    '故县':26,

}

tmp = od_by_area.drop('小区名称', axis=1)
tmp = tmp.sum()
tmp = tmp.sort_values(ascending=False)
tmp = tmp.reset_index()
tmp.columns = ['小区名称', '出行量']
tmp['小区编号'] = tmp['小区名称'].map(area_index)
tmp = tmp[['小区编号', '小区名称', '出行量']]

tmp.to_csv('小区总出行量.csv', index=False)
```

```python
# 进行聚类分析
from sklearn.cluster import KMeans
import numpy as np

# 除了"小区名称"列，其他列都是特征
X = od_by_area.drop('小区名称', axis=1)

# 选择聚类数目
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters)
kmeans.fit(X)

# 将聚类结果添加到原始数据中
od_by_area['cluster'] = kmeans.labels_

# 查看每个类别的小区数量
od_by_area['cluster'].value_counts()

# 打印各个类别的小区名称
for i in range(n_clusters):
    print('cluster', i)
    print(od_by_area[od_by_area['cluster'] == i]['小区名称'])
```
