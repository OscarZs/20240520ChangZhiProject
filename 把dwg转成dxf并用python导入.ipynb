```python
import ezdxf

# Load a DXF file
doc = ezdxf.readfile("Road Network and Traffic Zones of Changzhi City 2024.dxf")

# Access the model space, where most graphical objects are located
msp = doc.modelspace()

# Iterate over all graphic objects
for entity in msp:
    entity_type = entity.dxftype()
    print(f"Type: {entity_type}")

    # Handle different entity types appropriately
    if entity_type == 'LINE':
        print(f"  Start point: {entity.dxf.start}, End point: {entity.dxf.end}")
    elif entity_type == 'LWPOLYLINE' or entity_type == 'POLYLINE':
        # For polylines, print all vertices
        points = entity.get_points(format='xy')
        print(f"  Points: {points}")
    elif entity_type in ['CIRCLE', 'ARC']:
        print(f"  Center: {entity.dxf.center}, Radius: {entity.dxf.radius}")
        if entity_type == 'ARC':
            print(f"  Start angle: {entity.dxf.start_angle}, End angle: {entity.dxf.end_angle}")
    else:
        print("  Specific data for this entity type not handled in this script.")
```
